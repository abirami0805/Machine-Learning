```python
inputs = df.drop(['PassengerId','Name','SibSp','Parch','Ticket','Cabin','Embarked'],axis=1)
target = df['Survived']
print(inputs.head())
print(target.head())
```

# importing LabelEncoder from sklearn so as to convert text into numbers

```python
from sklearn.preprocessing import LabelEncoder
```

```python
sex = LabelEncoder()
age = LabelEncoder()
fare = LabelEncoder()
```

# adding the encoded columns to the inputs dataset

```python
inputs['sex'] = sex.fit_transform(inputs['Sex'])
inputs['age'] = sex.fit_transform(inputs['Age'])
inputs['fare'] = sex.fit_transform(inputs['Fare'])
inputs.head()
```

# dropping cells containing text and float values from inputs dataset such that we get a dataset with only numbers 

```python
input_n = inputs.drop(['Sex','Age','Fare'],axis=1)
input_n.head()
```

```python
target.head()
```

# import tree from sklearn and fit it into the model

```python
from sklearn import tree
model = tree.DecisionTreeClassifier()
model.fit(input_n,target)
```

# training the model 

```python
model.score(input_n,target)
```

# user values for prediction

```python
model.predict([[0,3,4,2,1]])
```
